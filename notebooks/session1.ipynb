__Shifting image down and right__

Translation matrix

shift 25 pixels on x-axis, 50 on y 

```python
M = np.float64([[1, 0, 25],
                [0, 1, 50]])
```

```python
width = img.shape[1]
height = img.shape[0]
```

```python
shifted = cv2.warpAffine(img, M, (width, height))
```

```python
jimshow(shifted, "Shifted down and right")
```

Shift 100 pixels left, 100 pixels up

```python
M = np.float64([[1, 0, -100],
                [0, 1, -100]])
```

```python
shifted = cv2.warpAffine(img, M, (width, height))
```

```python
jimshow(shifted, "Shifted up and left")
```

__What does this do?__

```python
M = np.float64([[1, 0, 0],
                [0, 1, 100]])
```

```python
shifted = cv2.warpAffine(img, M, (width, height))
```

```python
jimshow(shifted, "Shifted down")
```

```python
def translate(image, x, y):
    M = np.float64([[1, 0, x],
                [0, 1, y]])
    shifted = cv2.warpAffine(image, M, (width, height))
    
    return shifted
```

```python
shifted_img = translate(img, x=0, y=-100)
```

```python
jimshow(shifted_img)
```
